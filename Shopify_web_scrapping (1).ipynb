```python
import requests
from bs4 import BeautifulSoup
import lxml

O_Product_name = []
O_Retail_price = []
O_Actual_price = []
O_Discount = []
O_No_of_Ratings = []
O_Rating = []
O_RAM = []
O_ROM = []
O_Brand = []
```

```python
for j in range(42):
    
    Product_name = []
    Retail_price = []
    Actual_price = []
    Discount = []
    Rating = []
    RAM = []
    ROM = []
    Brand = []
    Rating = []
    
    url ="https://www.shopsy.in/search?q=smart+phones&page="+str(j)
    r = requests.get(url)
#   print(r)
    
    soup = BeautifulSoup(r.text, "lxml")
    # print(soup)
    box = soup.find("div", class_ = "sc-10414733-0 iGWAxy")
    
    
    names = box.find_all("div",class_ = "sc-c50e187b-0 bkNEtl")
    for i in names:
        name = i.text
        Product_name.append(name)
        brand = name.split()[0] 
        brand = brand.lower()
        Brand.append(brand)

    for product in Product_name:
        rom_part = product.split(",")[-1].strip()  
        rom = rom_part.replace(')', '')
        ROM.append(rom)

    ram = box.find_all("div",class_ = "css-146c3p1 r-dnmrzs r-1udh08x r-1udbk01 r-3s2u2q r-1iln25a r-1h7g6bg r-1et8rh5 r-1enofrn")
    for i in ram:
        rammm = i.text
        ram = rammm.replace('RAM', '')
        RAM.append(ram)

    cd =[]
    discount = box.find_all("div",class_ = "css-146c3p1 r-183gjk9 r-1vgyyaa r-1enofrn r-1rsjblm r-11wrixw")
    for i in discount:
        discount = i.text
        cd.append(discount)
    Discount = [discount.replace('% off', '').strip() for discount in cd]

    ap = box.find_all("div",class_ = "css-146c3p1 r-1h7g6bg r-1vgyyaa r-1enofrn r-1rsjblm r-142tt33 r-11wrixw")
    for i in ap:
        ap = i.text
        app = ap.replace(',','')
        Actual_price.append(app)
    
    if(len(Product_name) == len(Brand)== len(RAM)== len(Discount)==len(Actual_price)==len(ROM)):
        O_Product_name +=Product_name
        #O_Retail_price += Retail_price
        O_Actual_price += Actual_price
        O_Discount+=Discount
        #O_Rating+=Rating
        O_RAM+=RAM
        O_ROM+=ROM
        O_Brand+=Brand
```

```python
import pandas as pd
df = pd.DataFrame({"Product Name":O_Product_name,"Brand":O_Brand,"RAM":O_RAM,"ROM":O_ROM,"Discount(%)":O_Discount,"Price(Rs)":O_Actual_price})
```
